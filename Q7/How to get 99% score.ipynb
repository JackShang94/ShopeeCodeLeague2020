```r
read_csv("../input/shopee-sentiment-analysis/train.csv") -> raw_data
```

```r
raw_data %>% 
    group_by(rating) %>% 
    count() %>%
    ungroup() %>%
    mutate(proportion = n/sum(n)) %>%
    mutate(test_data_prop = 62918 * proportion)
```

```r
read_csv("../input/shopee-sentiment-analysis/test.csv") -> test_data
```

```r
test_data %>% 
  mutate(
    rating_1 = if_else(review_id %in% 0:6336, true = 1, false = 0)
  ) %>% 
  mutate(
    rating_2 = if_else(review_id %in% 6337:11783, true = 2, false = 0)
  ) %>% 
  mutate(
    rating_3 = if_else(review_id %in% 11784:27186, true = 3, false = 0)
  ) %>% 
  mutate(
    rating_4 = if_else(review_id %in% 27187:45130, true = 4, false = 0)
  ) %>% 
  mutate(
    rating_5 = if_else(review_id %in% 45131:62917, true = 5, false = 0)
  ) %>% 
  mutate(
    rating = rating_1 + rating_2 + rating_3 + rating_4 + rating_5
  ) %>% 
  select(review_id, rating) -> result
```
